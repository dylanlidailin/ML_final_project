```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

# Load the dataset
try:
    df = pd.read_csv("../Data/survey-lung-cancer.csv")
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: File not found. Please check the path '../Data/survey-lung-cancer.csv'")

# Display first few rows to verify
display(df.head())
```

```python
print("--- Dataset Info ---")
print(df.info())

print("\n--- Missing Values ---")
print(df.isnull().sum())

print("\n--- Summary Statistics (Numerical) ---")
display(df.describe())

print("\n--- Summary Statistics (Categorical) ---")
display(df.describe(include=['O']))
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='LUNG_CANCER', data=df, palette='viridis')
plt.title('Distribution of Target Variable (Lung Cancer)')
plt.xlabel('Lung Cancer Diagnosis')
plt.ylabel('Count')

# Add labels on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), 
                textcoords='offset points')
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# Histogram of Age
plt.subplot(1, 2, 1)
sns.histplot(df['AGE'], kde=True, bins=20, color='skyblue')
plt.title('Overall Age Distribution')

# Boxplot of Age vs Lung Cancer
plt.subplot(1, 2, 2)
sns.boxplot(x='LUNG_CANCER', y='AGE', data=df, palette='coolwarm')
plt.title('Age Distribution by Cancer Diagnosis')

plt.tight_layout()
plt.show()
```

```python
# Create a copy for correlation analysis to avoid modifying original data for other plots
df_corr = df.copy()

# Encode Binary Categorical Columns
df_corr['LUNG_CANCER'] = df_corr['LUNG_CANCER'].map({'YES': 1, 'NO': 0})
df_corr['GENDER'] = df_corr['GENDER'].map({'M': 1, 'F': 0})

# Calculate Correlation
plt.figure(figsize=(14, 10))
correlation_matrix = df_corr.corr()

# Draw Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
# Select a few key features to visualize
features_to_plot = ['SMOKING', 'ANXIETY', 'CHRONIC DISEASE', 'COUGHING']

plt.figure(figsize=(14, 10))

for i, feature in enumerate(features_to_plot, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=feature, hue='LUNG_CANCER', data=df, palette='Set2')
    plt.title(f'{feature} vs Lung Cancer')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.legend(title='Lung Cancer', loc='upper right')

plt.tight_layout()
plt.show()
```
